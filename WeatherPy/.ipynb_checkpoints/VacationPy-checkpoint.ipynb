### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
vacation = pd.read_csv("weather_data_df.csv")
vacation_df = pd.DataFrame(vacation)
vacation_df.head(10)
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
```

```python
fig = gmaps.figure()

locations = vacation_df[["Lat", "Lng"]].astype(float)

heat_layer = gmaps.heatmap_layer(locations, weights=vacation_df["Humidity"], 
                                 dissipating = False, max_intensity=100,
                                 point_radius = 2)

fig.add_layer(heat_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Temperature between 80 and 90
# Wind speed < 20
# Cloudiness < 30
# Humidity < 50

ideal_df = vacation_df.loc[vacation_df["Max Temp"] >= 80.00, :]
ideal_df = ideal_df.loc[ideal_df["Max Temp"] <= 90, :]
ideal_df = ideal_df.loc[ideal_df["Wind Speed"] < 20, :]
ideal_df = ideal_df.loc[ideal_df["Cloudiness"] < 30, :]
ideal_df = ideal_df.loc[ideal_df["Humidity"] < 50, :]
ideal_df
```
